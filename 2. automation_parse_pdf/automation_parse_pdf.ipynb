# 1. pdf berbasis text

## 1.1 Menggunakan Fitz

```python
def extract_text_from_pdf(pdf_path):
    """
    Menguraikan teks dari file PDF
    """
    doc = fitz.open(pdf_path)
    text = ""
    for page_num in range(doc.page_count):
        page = doc[page_num]
        text += page.get_text()

    return text

def main():
    pdf_files = pdf_text

    text_contents = []
    for pdf_file in pdf_files:
        text_content = extract_text_from_pdf(pdf_file)
        text_contents.append(text_content)
        print(f"Isi teks dari {pdf_file}:\n{text_content}\n{'='*30}\n")
    
    return text_contents

if __name__ == "__main__":
    result_text_content = main()
```

# 2. PDF yang mengandung tabel didalamnya

penggunaan library pada python yang digunakan untuk membaca tabel pada pdf sangat tergantung pada tabel yang ada pada file pdf nya. Bila tabel tersebut bersih dan memiliki struktur yang jelas, maka tabel yang dihasilkan akan rapih dan dapat langsung digunakan, namun bila strukturnya terlalu bervariasi, maka penggunaan library python masih cukup kesulitan untuk pengerjaan ini dan tabelnya masih perlu di bersihkan lagi,

Contoh pembacaan tabel menggunakan tabula (salah satu library pada python yang digunakan untuk mendeteksi tabel pada pdf) untuk mendeteksi tabel yang termasuk pada golongan sulit untuk dideteksi oleh library dasar python.

```python
# Tentukan path ke file PDF
file_path = "pdf_tabel.pdf"

# Gunakan tabula untuk membaca tabel dari PDF
tables = tabula.read_pdf(file_path, pages="all")
```

```python
table_1 = tables[0]
```

```python
table_1
```

masih terdapat beberapa missing value, hal tersebut karena pada pdf terdapat tabel yang barisnya tidak sama. Pada tabel, dimaksudkan adanya suatu kolom yang isinya mewakili lebih dari satu baris. Misalnya pada Phase EE, inforamsi tersebut digunakan untuk mewakili No 1.0 dan 2.0

```python
table_1_firsthalf = table_1[['No', 'Key Component', 'Score', 'Grade']]
table_1_secondhalf = table_1[['No.1', 'Key Component.1', 'Score.1', 'Grade.1']]
```

```python
table_1_firsthalf.dropna(inplace=True)
table_1_secondhalf.dropna(inplace=True)
table_1_firsthalf
```
